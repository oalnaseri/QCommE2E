# 01 - Channel Model Tutorial

This notebook demonstrates how to use the built-in quantum channel models.

```python
import numpy as np
from quantum_comm_sim.channels.bosonic_channel import BosonicChannel
from quantum_comm_sim.channels.depolarizing_channel import DepolarizingChannel
from quantum_comm_sim.utils.state_utils import ket_to_density
```

```python
psi = np.array([1, 0])
rho = ket_to_density(psi)
ch = BosonicChannel(dim=2, loss_db=3.0)
rho_out = ch.apply(rho)
print(rho_out)
```
